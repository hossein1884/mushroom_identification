```python
import numpy as np
import matplotlib.pyplot  as plt
```

```python
data=np.genfromtxt("mushrooms.csv",delimiter=",", dtype=str)
data=data[1:,[0,4,6,7,8,10]]
ms=np.zeros(np.shape(data))
# class: p=0 , e=1
ms[:,0][data[:,0]=='p']=0
ms[:,0][data[:,0]=='e']=1
# bruises: t=0 , f=1
ms[:,1][data[:,1]=='t']=0
ms[:,1][data[:,1]=='f']=1
# gill-attachment: f=0 , a=1
ms[:,2][data[:,2]=='f']=0
ms[:,2][data[:,2]=='a']=1
# gill-spacing: w=0 , c=1
ms[:,3][data[:,3]=='w']=0
ms[:,3][data[:,3]=='c']=1
# gill-size: n=0 , b=1
ms[:,4][data[:,4]=='n']=0
ms[:,4][data[:,4]=='b']=1
# stalk-shape: e=0 , t=1
ms[:,5][data[:,5]=='e']=0
ms[:,5][data[:,5]=='t']=1
ms
```

```python
def gini(x,y):
    arr=np.array([x,y]).T
    
    T=arr[arr[:,0]==1]
    F=arr[arr[:,0]==0]
    
    b1=len(T[T[:,1]==1])/len(T)
    b0=len(T[T[:,1]==0])/len(T)
    gini_T=1-(b1**2)-(b0**2)
    
    b1=len(F[F[:,1]==1])/len(F)
    b0=len(F[F[:,1]==0])/len(F)
    gini_F=1-(b1**2)-(b0**2)
    
    return (gini_T*(len(T)/len(y)))+(gini_F*(len(F)/len(y)))

print(f"ave bruises= {gini(ms[:,1],ms[:,0])}")
print(f"ave gill attachment= {gini(ms[:,2],ms[:,0])}")
print(f"ave gill spacing= {gini(ms[:,3],ms[:,0])}")
print(f"ave gill size= {gini(ms[:,4],ms[:,0])}")
print(f"ave stalk shape= {gini(ms[:,5],ms[:,0])}")
```
